```python
import pandas as pd
```

```python
df1=pd.read_csv('agano.csv')
df2=pd.read_csv('new_test.csv')
df1.columns=['index','content']
df2.columns=['index','content']
```

```python
def preprocessing(dataframe,name):
    dataframe.columns=['index','content']
    df=dataframe.drop(columns=['index'])
    df.reset_index(drop=True)
    df.to_csv(name,index=False)
    df=dataframe.iloc[:3235]
    return df
```

```python
import re
def clean_text(text):
    cleaned_text = re.sub(r'[\(\)\[\]\,\:\.\'\"0-9]|[\n]', ' ', text)  # Remove brackets, numbers, and replace \n
    return re.sub(r'\s+', ' ', cleaned_text).strip()       # Collapse spaces and strip whitespace
```

```python
df1['content']=df1['content'].apply(clean_text)
df2['content']=df2['content'].apply(clean_text)
```

```python
swahili_df=preprocessing(df1,'swahili.csv')
english_df=preprocessing(df2,'english.csv')
```

```python
final_df=pd.concat([swahili_df,english_df],axis=1)
```
