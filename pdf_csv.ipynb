```python
import PyPDF2
import re
```

```python
def extract_text_from_pdf(pdf_path,pattern):
    text=''
    pdf_file=open(pdf_path,'rb')
    pdf_reader=PyPDF2.PdfReader(pdf_file)
    num_pages=len(pdf_reader.pages)

    for page_num in range(num_pages):
        page=pdf_reader.pages[page_num]
        text+=page.extract_text()
    match = pattern.findall(text)
    df = pd.DataFrame({'content':match},index=[f'{i}' for i in range(len(match))])
    return df
```

```python
import pandas as pd
english_pdf='new_testament.pdf'
swahili_pdf='agano .pdf'
english_pattern = re.compile(r"(\d+:\d+)\s+(.+?)(?=\d+:\d+|$)", re.DOTALL)
swahili_pattern = re.compile(r"(\d+.\d+)\s+(.+?)(?=\d+.\d+|$)", re.DOTALL)
df=extract_text_from_pdf(english_pdf,english_pattern)
df2=extract_text_from_pdf(swahili_pdf,swahili_pattern)
```
